## 1.Indexing and Slicing Vector ndarrays

### 1-1.Indexing

```python
import numpy as np

a = np.arange(6)
print(f'ndarray:\n{a}\n')

print(f'a[0]: {a[0]}')
print(f'a[3]: {a[3]}')
print(f'a[-1]: {a[-1]}')
print(f'a[-3]: {a[-3]}\n')

# 리스트와 동일하게 for문 사용가능
for data in a:
    print(data)
```

### 1-2.Slicing

```python
import numpy as np

a = np.arange(10)
print(f'ndarray:\n{a}\n')

print(f'a[0:3]: {a[0:3]}')
print(f'a[3:7]: {a[3:7]}')
print(f'a[5:-1]: {a[5:-1]}\n')

print(f'a[2:]: {a[2:]}')
print(f'a[-2:]: {a[-2:]}\n') # 마지막 2개의 원소

print(f'a[:5]: {a[:5]}')
print(f'a[:-3]: {a[:-3]}')
print(f'a[:]: {a[:]}\n')

print(f'a[2:7:2]: {a[2:7:2]}')
print(f'a[::3]: {a[::3]}')
print(f'a[7:2:-1]: {a[7:2:-1]}')
print(f'a[::-1]: {a[::-1]}')
print(f'a[8:3:-2]: {a[8:3:-2]}')
print(f'a[::-3]: {a[::-3]}\n')

print(f'a[[0,3,6,-1]]: {a[[0,3,6,-1]]}')
print(f'a[a%3==0]: {a[a%3==0]}')
```

```python
import numpy as np

a = np.arange(1, 11)
print(f'ndarray:\n{a}\n')

a[:5] = 0
print(f'ndarray:\n{a}\n')

a[[6,8]] = 100
print(f'ndarray:\n{a}\n')

a[::2] = 200
print(f'ndarray:\n{a}\n')

a[5:-1:3] = 55
print(f'ndarray:\n{a}')
```

## 2.Indexing and Slicing Matrix ndarrays

```python
import numpy as np

a = np.arange(12).reshape((3,4))
print(f'ndarray:\n{a}\n')

print(f'a[0]: {a[0]}')
print(f'a[1]: {a[1]}')
print(f'a[2]: {a[2]}\n')

print(f'a[0,0], a[0,1], a[0,2]: {a[0,0], a[0,1], a[0,2], a[0,3]}')
print(f'a[1,0], a[1,1], a[1,2]: {a[1,0], a[1,1], a[1,2], a[1,3]}')
print(f'a[2,0], a[2,1], a[2,2]: {a[2,0], a[2,1], a[2,2], a[2,3]}\n')

print(f'a[0,2:]: {a[0,2:]}')
print(f'a[1,:3]: {a[1,:3]}')
print(f'a[2,1:3]: {a[2,1:3]}\n')

print(f'a[1:,0]: {a[1:,0]}')
print(f'a[:3,1]: {a[:2,1]}')
print(f'a[1:3,2]: {a[1:3,2]}')
```

```python
import numpy as np

a = np.arange(16).reshape((4,4))
print(f'ndarray:\n{a}\n')

print(f'a[1:3,1:3]:\n{a[1:3,1:3]}\n')
print(f'a[:2,:3]:\n{a[:2,:3]}\n')
print(f'a[-3:,-2:]:\n{a[-3:,-2:]}\n')
print(f'a[2:,:-2]:\n{a[2:,:-2]}')
```

```python
import numpy as np

image = np.arange(9).reshape((3,3))
print(f'ndarray:\n{image}\n')

horizontal_flip = image[:, ::-1]
print(f'horizontal_flip:\n{horizontal_flip}\n')

vertical_flip = image[::-1, :]
print(f'vertical_flip:\n{vertical_flip}\n')

rotation = image[::-1, ::-1]
print(f'rotation:\n{rotation}')
```

```python
import numpy as np

a = np.arange(16).reshape((4,4))
print(f'ndarray:\n{a}\n')

print(f'a[0, :]: {a[0, :]}') # 해당 차원을 모두 가져옴
print(f'a[0, ...]: {a[0,...]}\n') # 나머지 차원을 모두 가져옴

print(f'a[:, 1]: {a[:, 1]}')
print(f'a[..., 1]: {a[..., 1]}')
```

## 3.Indexing and Slicing 3rd Order Tensor ndarrays

```python
import numpy as np

a = np.arange(27).reshape((3,3,3))
print(f'ndarray:\n{a}\n')

print(f'a[0]:\n{a[0]}')
print(f'a[1]:\n{a[1]}')
print(f'a[2]:\n{a[2]}\n')

print(f'a[0,0]: {a[0,0]}')
print(f'a[0,1]: {a[0,1]}')
print(f'a[0,2]: {a[0,2]}')
print(f'a[1,0]: {a[1,0]}')
print(f'a[1,1]: {a[1,1]}')
print(f'a[1,2]: {a[1,2]}')
print(f'a[2,0]: {a[2,0]}')
print(f'a[2,1]: {a[2,1]}')
print(f'a[2,2]: {a[2,2]}\n')

print(f'a[0,:2,:2]:\n{a[0,:2,:2]}')
print(f'a[1,:2]:\n{a[1,:2]}')
print(f'a[2,-2:,-2:]:\n{a[2,-2:,-2:]}')
```

```python
import numpy as np

images = np.random.normal(size=(32, 100, 200))
print(f'image set: {images.shape}\n')

print(f'images[0,:,:].shape: {images[0,:,:].shape}')
print(f'images[0,...].shape: {images[0,...].shape}\n')

print(f'images[:,:,0].shape: {images[:,:,0].shape}')
print(f'images[...,0].shape: {images[...,0].shape}\n')

print(f'images[0,:,0].shape: {images[0,:,0].shape}')
print(f'images[0,...,0].shape: {images[0,...,0].shape}')
```

```python
import numpy as np

a = np.arange(27).reshape((3,3,3))
print(f'ndarray:\n{a}\n')

print(f'a[:,:2,:]:\n{a[:,:2,:]}\n')

print(f'a[:,:,0]:\n{a[:,:,0]}\n')

print(f'a[1:,-2:,-2:]:\n{a[1:,-2:,-2:]}')
```

```python
import numpy as np

image  = np.random.normal(size=(3,500,300))

image_r, image_g, image_b = image   # using unpacking

print(f'image_r.shape: {image_r.shape}')
print(f'image_g.shape: {image_g.shape}')
print(f'image_b.shape: {image_b.shape}\n')

top_left = image[:, :100, :100]
top_right = image[:, :100, -100:]
bottom_left = image[:, -100:, :100]
bottom_right = image[:, -100:, -100:]

print(f'top_left.shape: {top_left.shape}')
print(f'top_right.shape: {top_right.shape}')
print(f'bottom_left.shape: {bottom_left.shape}')
print(f'bottom_right.shape: {bottom_right.shape}')
```

## 4.Indexing with int ndarrays

### 4-1.Indexing and Slicing Vector ndarrays

```python
import numpy as np

a = np.arange(10)
print(f'ndarray:\n{a}\n')

print(f'a[[0,3,4]]: {a[[0,3,4]]}')
print(f'a[[0,0,5,5]]: {a[[0,0,5,5]]}\n')

indices = np.array([[1,2,3], [5,8,9]])
print(f'indices:\n{indices}')
print(f'a[indices]:\n{a[indices]}')
```

```python
import numpy as np

a = np.random.randint(0,20, (10,))
print(f'ndarray:\n{a}\n')

indices = np.random.randint(0,10, size=(2,3,4))
print(f'indices:\n{indices}')
print(f'a[indices]:\n{a[indices]}')
```

### 4-2.Indexing and Slicing Matrix ndarrays

```python
import numpy as np

a = np.arange(12).reshape((3,4))
print(f'nodarray: \n{a}\n')

indices= np.array([0,2])
print(f'indices:\n{indices}')
print(f'a[indices]:\n{a[indices]}\n')

indices = np.array([0,0,1,1,2,2])
print(f'indices:\n{indices}')
print(f'a[indices]:\n{a[indices]}\n')

indices = np.array([[0,1,2], [-3,-2,-1]])
print(f'indices:\n{indices}')
print(f'a[indices]:\n{a[indices]}\n')

indices0, indices1 = np.array([0,1,2]), np.array([1,2,3])
print('Paired indices:')
for idx0, idx1 in zip(indices0, indices1):
    print(f'{idx0, idx1}')
print(f'a[indices0, indices1]: {a[indices0, indices1]}')

indices0 = np.array([[0,1,2],[0,1,2]])
indices1= np.array([[0,1,2],[1,2,3]])
for row_indices0, row_indices1 in zip(indices0, indices1):
    for idx0, idx1 in zip(row_indices0, row_indices1):
        print(f'{idx0,idx1}',end=' ')
    print()
print(f'a[indices0, indices1]:\n{a[indices0, indices1]}')
```

### 5-3.Extracting Indices

```python
import numpy as np

a = np.array([True, False, True, False])

nonzero = np.nonzero(a)
where = np.where(a)
print(f'a:\n{a}')
print(f'nonzero:\n{nonzero}')
print(f'where:\n{where}\n')

a = np.array([[True,False],
              [True, False]])

nonzero = np.nonzero(a)
where = np.where(a)
print(f'a:\n{a}')
print(f'nonzero:\n{nonzero}')
print(f'where:\n{where}')

a = np.array([[[True, False, True],
              [True, False, False]],
              
              [[False, True, False],
               [True, False, True]]])

nonzero = np.nonzero(a)
where = np.where(a)
print(f'a:\n{a}\n')
print(f'nonzero:\n{nonzero}')
print(f'where:\n{where}')
```

```python
import numpy as np

a = np.random.randint(-2,3, size=(3,3))

u_nonzero = a[np.nonzero(a)]
u_where = a[np.where(a)]
u_bool = a[a != 0]

print(f'a:\n{a}\n')

print(f'using nonzero: \n{u_nonzero}')
print(f'using where: \n{u_where}')
print(f'using bool ndarray: \n{u_bool}\n')

u_nonzero = a[np.nonzero(a > 0)]
u_where = a[np.where(a > 0)]
u_bool = a[a > 0]

print(f'using nonzero: \n{u_nonzero}')
print(f'using where: \n{u_where}')
print(f'using bool ndarray: \n{u_bool}')
```
