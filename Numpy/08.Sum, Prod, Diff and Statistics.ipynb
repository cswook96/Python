## 1.ndarray의 합곱차 구하기


### 1-1.summation
- np.cumsum(a, axis=None, dtype=None, out=None)
- ndarray.cumsum(a, axis=None, dtype=None, out=None)

```python
import numpy as np

a = np.arange(5)
print(f'ndarray: {a}\n')

cumsum = np.cumsum(a)
print(f'cumsum: {cumsum}')
```

```python
import numpy as np

a = np.arange(12).reshape((3,-1))
print(f'ndarray:\n{a}, shape: {a.shape}\n')

cumsum = np.cumsum(a)
print(f'cumsum:\n{cumsum}, shape: {cumsum.shape}\n')

cumsum = np.cumsum(a, axis=0)
print(f'cumsum(axis=0):\n{cumsum}, shape: {cumsum.shape}\n')

cumsum = np.cumsum(a, axis=1)
print(f'cumsum(axis=1):\n{cumsum}, shape: {cumsum.shape}')
```

### 1-2.Product
- np.prod(a, axis=None, dtype=None, out=None, keepdims=no value, initial=no value, where=no value)
- np.cumprod(a ,axis=None, dtype=None, out=None)

```python
import numpy as np

a = np.arange(1, 5)
print(f'ndarray: {a}\n')

prod = np.prod(a)
cumprod = np.cumprod(a)
print(f'prod: {prod}')
print(f'cumprod: {cumprod}, shape: {cumprod}')
```

```python
import numpy as np

a = np.arange(1,13).reshape((3,-1))
print(f'ndarray:\n{a}, shape: {a.shape}\n')

prod = np.prod(a, axis=0)
cumprod = np.cumprod(a, axis=0)
print(f'prod(axis=0):\n{prod}, shape: {prod.shape}\n')
print(f'cumprod(axis=0):\n{cumprod}, shape: {cumprod.shape}\n')

prod = np.prod(a, axis=1, keepdims=True)
cumprod = np.cumprod(a, axis=1)
print(f'prod(axis=1, keepdims=True):\n{prod}, shape: {prod.shape}\n')
print(f'cumprod(axis=1):\n{cumprod}, shape: {cumprod.shape}')
```

### 1-3.Difference

- ndarray.mean(axis=None, dtype=None, out=None, keepdims=no value, *, where=no value)

### 2-3.Variance and Standard Deviation
- np.var(a, axis=None, dtype=None, out=None, ddof=0, keepdims=no value, *,where=no value)
- ndarray.var(axis=None, dtype=None, out=None, ddof=0, keepdims=no value, *,where=no value)
- np.std(a, axis=None, dtype=None, out=None, ddof=0, keepdims=no value, *,where=no value)
- ndarray.std(axis=None, dtype=None, out=None, ddof=0, keepdims=no value, *,where=no value)

```python
import numpy as np

scores = np.random.normal(10,5, (100,))

var = scores.var()
std = scores.std()

print(f'variance: {var}')
print(f'standard deviation: {std}')
```

### 2-4.Standardization

```python
import numpy as np

means = [50,60,70]
stds = [3,5,10]
n_student, n_class = 100, 3

scores = np.random.normal(loc=means, scale=stds,
                          size=(n_student,n_class))
scores = scores.astype(np.float32)


mean = np.mean(scores, axis=0)
std = np.std(scores, axis=0)
print(f'Shape of scores: {scores.shape}, dtype: {scores.dtype}')
print(f'Mean: {mean}, shape = {mean.shape}')
print(f'Std: {std}, shape = {std.shape}\n')

score_stdz = (scores - mean)/std
mean = np.mean(score_stdz, axis=0)
std = np.std(score_stdz, axis=0)
print(f'Shape of scores: {score_stdz.shape}')
print(f'Mean: {mean}, shape = {mean.shape}')
print(f'Std: {std}, shape = {std.shape}')
```

### 2-5.Max Values and Indices
- np.amax(a, axis=None, dtype=None, out=None, keepdims=no value, initial=no value, where=no value)
- ndarray.max(axis=None, dtype=None, out=None, keepdims=no value, initial=no value, where=no value)
- np.argmax(a, axis=None, out=None)
- ndarray.argmax(axis=None, out=None)

```python
import numpy as np

a = np.random.randint(0,100, (10,))
print(f'ndarray:\n{a}')

M = np.max(a)
M_idx = np.argmax(a)
print(f'M/M_idx: {M}/{M_idx}')
```

```python
import numpy as np

means = [50,60,70]
stds = [3,5,10]
n_student, n_class = 100, 3

scores = np.random.normal(loc=means, scale=stds,
                          size=(n_student,n_class))
scores = scores.astype(np.float32)
print(f'Shape of scores: {scores.shape}, dtype: {scores.dtype}\n')

scores_max = np.max(scores, axis=0)
scores_max_idx = np.argmax(scores, axis=0)
print(f'max scores(axis=0): {scores_max}, shape: {scores_max.shape}')
print(f'max indices(axis=0): {scores_max_idx}, shape: {scores_max_idx.shape}\n')

scores_max = np.max(scores, axis=1)
scores_max_idx = np.argmax(scores, axis=1)
print(f'Shape of max scores(axis=1): {scores_max.shape}')
print(f'Shape of max indices(axis=1): {scores_max_idx.shape}')
```

### 2-6.Min Values and Indices
- np.amin(a, axis=None, dtype=None, out=None, keepdims=no value, initial=no value, where=no value)
- ndarray.min(axis=None, dtype=None, out=None, keepdims=no value, initial=no value, where=no value)
- np.argmin(a, axis=None, out=None)
- ndarray.argmin(axis=None, out=None)

```python
# Min-max Normalization
import numpy as np

means = [50,60,70]
stds = [3,5,10]
n_student, n_class = 100, 3

scores = np.random.normal(loc=means, scale=stds,
                          size=(n_student,n_class))
scores = scores.astype(np.float32)
print(f'Shape of scores: {scores.shape}, dtype:{scores.dtype}')

M = np.max(scores, axis=0)
m = np.min(scores, axis=0)
print(f'max values: {M}, shape: {M.shape}')
print(f'min values: {m}, shape: {m.shape}\n')

score_nor = (scores - m)/(M-m)
print(f'Shape of Normalization: {score_nor.shape}')
M = np.max(score_nor, axis=0)
m = np.min(score_nor, axis=0)
print(f'Max Values: {M}, shape: {M.shape}')
print(f'min Values: {m}, shape: {m.shape}')
```

### 2-7.Maximun and Minimum

```python
import numpy as np

u = np.random.randint(0,10, (10,))
v = np.random.randint(0,10, (10,))
print(f'u:\n{u}, shape: {u.shape}')
print(f'v:\n{v}, shape: {v.shape}\n')

maximum = np.maximum(u, v)
print(f'maximum: {maximum}, shape: {maximum.shape}')
minimum = np.minimum(u, v)
print(f'minimum: {minimum}, shape: {minimum.shape}')
```

```python
import numpy as np

u = np.random.randint(0,10, (3,4))
v = np.random.randint(0,10, (3,4))
print(f'u:\n{u}, shape: {u.shape}')
print(f'v:\n{v}, shape: {v.shape}\n')

maximum = np.maximum(u, v)
print(f'maximum:\n{maximum}, shape: {maximum.shape}')
minimum = np.minimum(u, v)
print(f'minimum:\n{minimum}, shape: {minimum.shape}')
```

```python
import numpy as np

u = np.random.randint(0,10, (10,))
v = np.random.randint(0,10, (10,))
print(f'u:\n{u}, shape: {u.shape}')
print(f'v:\n{v}, shape: {v.shape}\n')

maximum = np.zeros_like(u)
maximum[u >= v] = u[u >= v]
maximum[v > u] = v[v > u]
print(f'maximum:\n{maximum}, shape: {maximum.shape}')

minimum = np.zeros_like(u)
minimum[u <= v] = u[u <= v]
minimum[v < u] = v[v < u]
print(f'minimum:\n{minimum}, shape: {minimum.shape}')
```

```python
import numpy as np

u = np.random.randint(0,10, (10,))
v = np.random.randint(0,10, (10,))
print(f'u:\n{u}, shape: {u.shape}')
print(f'v:\n{v}, shape: {v.shape}\n')

# u가 더크면 100, v가 더 크면 -100
up_vals = np.full_like(u, fill_value=100)
down_vals = np.full_like(u, fill_value=-100)
print(np.where(u>v, up_vals, down_vals))
```

```python
import numpy as np

u = np.random.randint(0,10, (10,))
v = np.random.randint(0,10, (10,))
print(f'u:\n{u}, shape: {u.shape}')
print(f'v:\n{v}, shape: {v.shape}\n')

maximum = np.maximum(u,v)
maximum_where = np.where(u>v,u,v)
print(f'maximum:\n{maximum}, shape: {maximum.shape}')
print(f'maximum(where): \n{maximum_where}, shape: {maximum_where.shape}\n')

minimum = np.minimum(u,v)
minimum_where = np.where(u<v,u,v)
print(f'minimum:\n{minimum}, shape: {minimum.shape}')
print(f'minimum(where): \n{minimum_where}, shape: {minimum_where.shape}')
```
