## 내림차순

```python
scores = [40,20,30,-20,-10,50,70]
print(f'scores:{scores}')
sorted_scores = []

for _ in range(len(scores)):
    M,M_idx = scores[0],0

    for score_idx in range(len(scores)):
        if scores[score_idx] > M:
            M = scores[score_idx]
            M_idx = score_idx
    
    remaing_scores = []
    for score_idx in range(len(scores)):
        if score_idx == M_idx:
            sorted_scores.append(M)
        else:
            remaing_scores.append(scores[score_idx])
    scores = remaing_scores
    print(f'remaing scores: {scores} ---> sorted scores: {sorted_scores}')
```

## 오름차순

```python
scores = [40,20,30,-20,-10,50,70]
print(f'scores:{scores}')
sorted_scores = []

for _ in range(len(scores)):
    m,m_idx = scores[0],0

    for score_idx in range(len(scores)):
        if scores[score_idx] < m:
            m = scores[score_idx]
            m_idx = score_idx
    
    remaing_scores = []
    for score_idx in range(len(scores)):
        if score_idx == m_idx:
            sorted_scores.append(m)
        else:
            remaing_scores.append(scores[score_idx])
    scores = remaing_scores
    print(f'remaing scores: {scores} ---> sorted scores: {sorted_scores}')
```
