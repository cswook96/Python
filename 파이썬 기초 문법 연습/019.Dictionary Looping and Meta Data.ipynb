## Dictionary Looping

```python
cnt_dict = {'a':1,'b':2,'c':3,'d':4}

for key in cnt_dict:
    print(f'key: {key}, value: {cnt_dict[key]}')
```

## dict.keys, dict.values, dict.items

```python
cnt_dict = {'a':1,'b':2,'c':3,'d':4}

print(cnt_dict.keys())
for k in cnt_dict.keys():
    print(k)

print('=====')

print(cnt_dict.values())
for v in cnt_dict.values():
    print(v)

print('=====')

print(cnt_dict.items())
for k, v in cnt_dict.items():
    print(f'key:{k}, value:{v}')
```

## 가장 많이 등장한 글자 찾기

```python
zen_of_python = \
'''
The Zen of Python, by Tim Peters

Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
Flat is better than nested.
Sparse is better than dense.
Readability counts.
Special cases aren't special enough to break the rules.
Although practicality beats purity.
Errors should never pass silently.
Unless explicitly silenced.
In the face of ambiguity, refuse the temptation to guess.
There should be one-- and preferably only one --obvious way to do it.
Although that way may not be obvious at first unless you're Dutch.
Now is better than never.
Although never is often better than *right* now.
If the implementation is hard to explain, it's a bad idea.
If the implementation is easy to explain, it may be a good idea.
Namespaces are one honking great idea -- let's do more of those!
'''

cnt_dict = {}

for char in zen_of_python:
    cnt_dict[char] = cnt_dict.get(char,0)+1
print('cnt_dict:\n',cnt_dict,'\n')

M_char,M_freq = None,None
for char, char_freq in cnt_dict.items():
    if M_freq == None or (char_freq > M_freq and char != ' '):
        M_char = char
        M_freq = char_freq
print(f'M_char: {M_char}, M_freq: {M_freq}')
```

## Score Meta Data Dictionary

### 한과목 점수들에 대한 평균,분산,표준편차

```python
scores = [50, 30, 90, 70, 30, 95, 40]
n_student = len(scores)

score_sum, score_square_sum = 0,0

for score in scores:
    score_sum += score
    score_square_sum += score**2

score_meta_data = {}
score_meta_data['mean'] = score_sum/n_student
score_meta_data['variance'] = score_square_sum/n_student - score_meta_data['mean']**2
score_meta_data['std'] = score_meta_data['variance']**0.5

template = 'Mean: {:.2f}\n' + \
            'Variance: {:.2f}\n' + \
            'Standard deviation: {:2f}'

print(template.format(score_meta_data['mean'],
                      score_meta_data['variance'],
                      score_meta_data['std']))
```

### 여러과목 합,평균,최고점,최고점 인덱스

```python
scores = [[10,15,30],
          [20,45,20],
          [30,35,40],
          [40,30,50]]

class_names = ['Korean','English','Math']
interested_data = ['sum','mean','M','M_idx']

n_studnet = len(scores)
n_class = len(scores[0])

meta_data_dict = {}
for class_name in class_names:
    meta_data_dict[class_name] = {}
    for i_data in interested_data:
        meta_data_dict[class_name][i_data] = 0

for s_idx, s_scores in enumerate(scores):
    for c_idx,c_score in enumerate(s_scores):
        meta = meta_data_dict[class_names[c_idx]]
        meta['sum'] += c_score

        if c_score > meta['M']:
            meta['M'] = c_score
            meta['M_idx'] = s_idx

for c_idx in range(n_class):
    meta = meta_data_dict[class_names[c_idx]]
    meta['mean'] = meta['sum']/n_studnet

for k, v in meta_data_dict.items():
    print(f'subject: {k} ==> {v}')
```
