## 평균점수 구하기

### 학생별 평균점수 구하기

```python
scores = [[10,15,20],[20,25,30],[30,35,40],[40,45,50]]
n_subject = len(scores[0])
student_means=[]


for student_scores in scores:
    subject_sum = 0
    for score in student_scores:
        subject_sum += score
    student_means.append(subject_sum/n_subject)

print('학생별 평균점수:',student_means)
```

### 과목별 평균점수 구하기

```python
scores = [[10,15,20],[20,25,30],[30,35,40],[40,45,50]]
n_student = len(scores)
n_subject = len(scores[0])
subject_sums = []
subject_means = []

for _ in range(n_subject):
    subject_sums.append(0)

for student_idx in range(n_student):
    for score_idx in range(n_subject):
        subject_sums[score_idx] += scores[student_idx][score_idx]

for sum in subject_sums:
    subject_means.append(sum/n_student)

print('과목별 평균점수: ',subject_means)
```

## Mean Subtraction

###  학생별 Mean Subtraction

```python
scores = [[10,15,20],[20,25,30],[30,35,40],[40,45,50]]
n_student = len(scores)
n_subject = len(scores[0])

# --- 학생별 평균,분산,표준편차 구하기
student_means=[]
student_variances = []
student_stds = []

for student_scores in scores:
    subject_sum = 0
    subject_square_sum = 0
    for score in student_scores:
        subject_sum += score
        subject_square_sum += score**2
    student_means.append(subject_sum/n_subject)
    student_variances.append(subject_square_sum/n_subject-(subject_sum/n_subject)**2)
    student_stds.append((subject_square_sum/n_subject-(subject_sum/n_subject)**2)**0.5)

print('학생별 평균 : ',student_means)
print('학생별 분산 : ',student_variances)
print('학생별 표준편차 : ',student_stds)
# ---

# --- 학생별 Mean Subtraction
for student_idx in range(n_student):
    for subject_idx in range(n_subject):
        scores[student_idx][subject_idx]-= student_means[student_idx]

print('Mean Subtraction를 한후 scores :', scores)
#---

# --- Mean Subtraction를 한후 학생별 평균,분산,표준편차
student_means=[]
student_variances = []
student_stds = []

for student_scores in scores:
    subject_sum = 0
    subject_square_sum = 0
    for score in student_scores:
        subject_sum += score
        subject_square_sum += score**2
    student_means.append(subject_sum/n_subject)
    student_variances.append(subject_square_sum/n_subject-(subject_sum/n_subject)**2)
    student_stds.append((subject_square_sum/n_subject-(subject_sum/n_subject)**2)**0.5)

print('Mean Subtraction를 한후 학생별 평균: ',student_means)
print('Mean Subtraction를 한후 학생별 분산: ',student_variances)
print('Mean Subtraction를 한후 학생별 표준편차: ',student_stds)
#---
```

###  과목별 Mean Subtraction

```python
scores = [[10,15,20],[20,25,30],[30,35,40],[40,45,50]]
n_student = len(scores)
n_subject = len(scores[0])

# ---과목별 평균,분산,표준편차 구하기
subject_sums = []
subject_square_sums = []
subject_means = []
subject_variances = []
subject_stds = []

for _ in range(n_subject):
    subject_sums.append(0)
    subject_square_sums.append(0)

for student_idx in range(n_student):
    for score_idx in range(n_subject):
        subject_sums[score_idx] += scores[student_idx][score_idx]
        subject_square_sums[score_idx] += scores[student_idx][score_idx]**2

for subject_idx in range(n_subject):
    subject_means.append(subject_sums[subject_idx]/n_student)
    subject_variances.append(subject_square_sums[subject_idx]/n_student-subject_means[subject_idx]**2)
    subject_stds.append(subject_variances[subject_idx]**0.5)

print('과목별 평균 : ',subject_means)
print('과목별 분산 : ',subject_variances)
print('과목별 표준편차 : ',subject_stds)
# ---

# --- 과목별 Mean Subtraction
for subject_idx in range(n_subject):
    for student_idx in range(n_student):
        scores[student_idx][subject_idx] -= subject_means[subject_idx]

print('Mean Subtraction를 한후 scores:', scores)
# ---

# --- Mean Subtraction를 한후 과목별 평균,분산,표준편차 
subject_sums = []
subject_square_sums = []
subject_means = []
subject_variances = []

for _ in range(n_subject):
    subject_sums.append(0)
    subject_square_sums.append(0)

for student_idx in range(n_student):
    for score_idx in range(n_subject):
        subject_sums[score_idx] += scores[student_idx][score_idx]
        subject_square_sums[score_idx] += scores[student_idx][score_idx]**2

for subject_idx in range(n_subject):
    subject_means.append(subject_sums[subject_idx]/n_student)
    subject_variances.append(subject_square_sums[subject_idx]/n_student-subject_means[subject_idx]**2)
    subject_stds.append(subject_variances[subject_idx]**0.5)

print('Mean Subtraction를 한후 과목별 평균 : ',subject_means)
print('Mean Subtraction를 한후 과목별 분산 : ',subject_variances)
print('Mean Subtraction를 한후 과목별 표준편차 : ',subject_stds)
#---
```

## Standardization

```python
scores = [[10,15,20],[20,25,30],[30,35,40],[40,45,50]]
n_student = len(scores)
n_subject = len(scores[0])

# ---과목별 평균,분산,표준편차 구하기
subject_sums = []
subject_square_sums = []
subject_means = []
subject_variances = []
subject_stds = []

for _ in range(n_subject):
    subject_sums.append(0)
    subject_square_sums.append(0)

for student_idx in range(n_student):
    for score_idx in range(n_subject):
        subject_sums[score_idx] += scores[student_idx][score_idx]
        subject_square_sums[score_idx] += scores[student_idx][score_idx]**2

for subject_idx in range(n_subject):
    subject_means.append(subject_sums[subject_idx]/n_student)
    subject_variances.append(subject_square_sums[subject_idx]/n_student-subject_means[subject_idx]**2)
    subject_stds.append(subject_variances[subject_idx]**0.5)

print('과목별 평균 : ',subject_means)
print('과목별 분산 : ',subject_variances)
print('과목별 표준편차 : ',subject_stds)
# ---

# ---과목별 Standardization
for subject_idx in range(n_subject):
    for student_idx in range(n_student):
        scores[student_idx][subject_idx] = (scores[student_idx][subject_idx]- subject_means[subject_idx])/subject_stds[subject_idx]

print('scores:', scores)
# ---

# --- Standardization를 한후 과목별 평균,분산,표준편차 
subject_sums = []
subject_square_sums = []
subject_means = []
subject_variances = []
subject_stds = []

for _ in range(n_subject):
    subject_sums.append(0)
    subject_square_sums.append(0)

for student_idx in range(n_student):
    for score_idx in range(n_subject):
        subject_sums[score_idx] += scores[student_idx][score_idx]
        subject_square_sums[score_idx] += scores[student_idx][score_idx]**2

for subject_idx in range(n_subject):
    subject_means.append(subject_sums[subject_idx]/n_student)
    subject_variances.append(subject_square_sums[subject_idx]/n_student-subject_means[subject_idx]**2)
    subject_stds.append(subject_variances[subject_idx]**0.5)

print('Standardization를 한후 과목별 평균 : ',subject_means)
print('Standardization를 한후 과목별 분산 : ',subject_variances)
print('Standardization를 한후 과목별 표준편차 : ',subject_stds)
# ---
```
