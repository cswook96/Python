## Accuracy

```python
predictions = [1,1,0,2,1,2,1]
labels = [1,1,0,3,1,2,1]
n_data = len(predictions)
n_correct = 0

for data_idx in range(n_data):
    if predictions[data_idx] == labels[data_idx]:
        n_correct += 1

accuracy = n_correct/n_data

print('Accuracy:',accuracy)
```

### 클래스별 Accuracy

```python
predictions = [0,1,0,2,1,2,0]
labels = [1,1,0,0,1,2,1]
n_data = len(labels)

# --- 클래스 갯수 구하기
M_class = labels[0]
for label_idx in range(n_data):
    if labels[label_idx] > M_class:
        M_class = labels[label_idx]
n_class = M_class + 1
# ---

class_cnts = [] # 각 클래스별 데이터의 수
correct_cnts = [] # 각 클래스별 예측이 맞은 데이터의 수
accuracy_vec = [] # 각 클래스별 정확도

for _ in range(n_class):
    class_cnts.append(0)
    correct_cnts.append(0)
    accuracy_vec.append(None)

for data_idx in range(n_data):
    pred = predictions[data_idx]
    label = labels[data_idx]

    class_cnts[label] += 1

    if pred == label:
        correct_cnts[label] += 1

for class_idx in range(n_class):
    accuracy_vec[class_idx] = correct_cnts[class_idx]/class_cnts[class_idx]

print('클래스별 정확도:',accuracy_vec)
```

## Histogram

```python
scores = [50,20,10,20,30,70,80,35,90,40,55]
cutoffs = [0,20,40,60,80,100] # 점수 구간 (0~20,20~40,40~60,60~80,80~100)
n_interval = len(cutoffs)-1
histogram = []   

# --- 구간별 도수 초기화(0~20,20~40,40~60,60~80,80~100)
for _ in range(n_interval):
    histogram.append(0)
# ---

for score in scores:
    for interval in range(n_interval,0,-1):
        if score >= cutoffs[interval-1] and score < cutoffs[interval]:
            histogram[interval-1] += 1
        else:
            pass

print('각 구간별 점수 도수: ',histogram)
```

## 절댓값 구하기

```python
numbers = [-4,1,-3,5,-2,-8,9]
abs_numbers = []

for number in numbers:
    if number < 0:
        abs_numbers.append(-number)
    else:
        abs_numbers.append(number)

print(f'numbers:{numbers} ---> abs_numbers:{abs_numbers}')
```

## Manhattan Distance

```python
v1 = [2,9,11,3,7,5,2]
v2 = [2,3,4,3,12,1,3]

m_distance = 0
for dim_idx in range(len(v1)):
    sub = v1[dim_idx] - v2[dim_idx]

    if sub < 0:
        m_distance += -sub
    else:
        m_distance += sub

print('Manhattan Distance: ',m_distance)
```
