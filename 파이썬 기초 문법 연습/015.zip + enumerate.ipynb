## 두 과목의 최댓값과 그값들의 index

```python
m_scores = [20,60,30,40]
e_scores = [40,20,50,80]

Maxes = [m_scores[0],e_scores[1]]
M_indices = [0,0]

for student_idx, (m_score, e_score) in enumerate(zip(m_scores,e_scores)):
    if m_score > Maxes[0]:
        Maxes[0] = m_score
        M_indices[0] = student_idx
    
    if e_score > Maxes[1]:
        Maxes[1] = e_score
        M_indices[1] = student_idx
    
print('두 과목의 최댓값:',Maxes)
print('최댓값 indices:',M_indices) 
```
