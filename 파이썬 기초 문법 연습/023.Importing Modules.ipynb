## random.randint

```python
import random

scores = [random.randint(0, 99) for _ in range(100)]
print('scores:\n',scores)

n_bin = 5
hist_range = [0, 100]
interval = (hist_range[1] - hist_range[0])/n_bin
cutoffs = []
hist = []

for bin in range(n_bin):
    cutoffs.append(hist_range[0] + interval*bin)
    hist.append(0)
print('\ncutoffs:\n',cutoffs)

for score in scores:
    for cutoff_idx,cutoff in enumerate(cutoffs):
        diff = score - cutoff
        if diff >=0 and diff < interval:
            hist[cutoff_idx] += 1
            break
print('\nhistogram:\n',hist)
```

## random.gauss

```python
import random

scores = [random.gauss(0,3)  # (평균,표준편차)
          for _ in range(100)]
print(scores)

n_bin = 10
hist_range = [- 10, 10]
interval = (hist_range[1] - hist_range[0])/n_bin
cutoffs = []
hist = []

for bin in range(n_bin):
    cutoffs.append(hist_range[0] + interval*bin)
    hist.append(0)
print('\ncutoffs:\n',cutoffs)

for score in scores:
    for cutoff_idx,cutoff in enumerate(cutoffs):
        diff = score - cutoff
        if diff >=0 and diff < interval:
            hist[cutoff_idx] += 1
            break

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(7,7))
ax.bar(cutoffs,hist,width=interval,edgecolor='skyblue',linewidth=3,align='edge')
ax.tick_params(labelsize=10)
```

## Math Libary

```python
import math

PI = math.pi
print('PI:',PI)

sin = math.sin(PI)
cos = math.cos(PI)
print('sin(PI[radian]):',sin)
print('cos(PI[radian]):',cos)
print('\n')

degree = 90
sin = math.sin(degree/180*PI)
cos = math.cos(degree/180*PI)
print('sin(90[degree]):',sin)
print('cos(90[degree]):',cos)
```

## Linear space

```python
low, high = 20,100
n_data = 11

interval = (high - low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]
print('linspace:',linspace)
print('length of linspace:',len(linspace))
```

## Matplotlib

```python
import matplotlib.pyplot as plt
import random

fig, axes = plt.subplots(3,1,figsize=(8,12))

# line plot
scores = [50,30,90,70]
axes[0].plot([4,6,8,10],scores)

# scatter plot
x_data = [random.gauss(0,1) for _ in range(100)]
y_data = [random.gauss(0,1) for _ in range(100)]
axes[1].scatter(x_data,y_data,alpha = 0.3)

#histogram
n_student = 10000
scores = [random.gauss(50,5) for _ in range(n_student)]
axes[2].hist(scores,bins=20)
```
