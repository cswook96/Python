## Usage

```python
names = ['A','B','C','D','E']
scores = [50,20,80,40,50]

data_dict = {name : score for name,score in zip(names,scores)}
print(data_dict)
```

## Dictionary to List(keys, values)

```python
name_score_dict = {'A':50,'B':30,'C':20,'D':40,'E':50}

student_names = ['Studnet ' + name for name in name_score_dict.keys()]
print(student_names)

doubled_scores = [2*score for score in name_score_dict.values()]
print(doubled_scores)
```

## Dictionary to Dictionary Comprehension

```python
 name_score_dict = {'A':50,'B':30,'C':20,'D':40,'E':50}

name_doubled_score_dict = {'Student ' + name : 2*score 
                           for name,score in name_score_dict.items()}
print(name_doubled_score_dict)
```

## Key,Value 바꾸기

```python
name_score_dict = {'A':50,'B':30,'C':20,'D':40,'E':50}

score_name_dict = {score : name 
                   for name,score in name_score_dict.items()}
print(score_name_dict)
```
