## Usage

```python
score1, score2 = 10, 20

template = 'score1: {}, score2: {}'
print(template.format(score1,score2))
```

## format의 위치 정하기

```python
score1, score2, score3 = 10, 20, 30

template = 'score1: {1}, score2: {0}, score3: {2}'
print(template.format(score1,score2, score3))
```

## 출력되는 소숫점 개수 정하기

```python
mean1 = 23.12313251
mean2 = 51.23124355

template = 'mean1: {:.2f}\nmean2: {:.2f}' # 소숫점 둘째 자리까지 표시
print(template.format(mean1,mean2))
```

## String Concatenation

```python
test_string = 'Choi ' + 'Sungwook'
print(test_string)

test_string = "I've got " + str(100) + ' point!' # string 과 숫자는 연산이 안되므로 숫자를 string으로 바꾸어 줘야한다
print(test_string)
```

## 여러줄에 걸친 string template

```python
mean1 = 23.233251
mean2 = 12.213325
mean3 = 64.232145

template = 'mean1: {:.2f}\n' + \
     'mean2: {:.2f}\n' + \
     'mean3: {:.2f}'
print(template.format(mean1,mean2,mean3))
```

## f String

```python
mean1, mean2 = 10.21312, 20.12124

print(f'mean1: {mean1:.2f}, mean2: {mean2:.2f}')
```

## 평균,분산,표준편차 구하기

```python
scores = [50,30,90,70,30,95,40]
n_student = len(scores)

sum,square_sum = 0,0

for score in scores:
    sum += score
    square_sum += score**2

mean = sum/n_student
variance = square_sum/n_student - mean**2
std = variance**0.5

# --- method 1
print('---- method 1 ----')

template = f'Score mean: {mean:.2f}\n' + \
    f'Score variance: {variance:.2f}\n' + \
    f'Score standard deviation: {std:.2f}\n'

print(template)
# ---

# --- method 2
print('---- method 2 ----')

template = 'Score mean: {:.2f}\n' + \
    'Score variance: {:.2f}\n' + \
    'Score standard deviation: {:.2f}'

print(template.format(mean,variance,std))
# ---
```
