## Comprehension + 연산

```python
# 2의배수들 만들기
mitiple_of_two = [2*i for i in range(20)]
print(mitiple_of_two)

# 3이 더해진수 만들기
added_by_three = [i * 3 for i in range(20)]
print(added_by_three)

# 제곱된수 만들기
squares = [num**2 for num in range(20)]
print(squares)
```

## Comprehension으로 List 접근하기

```python
scores = [20,60,20,10,70]

double_scores = [2*score for score in scores]
print(double_scores)
```

## Comprehension에서 Inedex로 여러개 list 접근하기

```python
m_scores = [10*i for i in range(10)]
e_scores = [8*i for i in range(10)]

score_sums = [m_scores[score_idx] + e_scores[score_idx] for score_idx in range(len(m_scores))]
print(score_sums)
```

## Comprehension + zip

```python
m_scores = [10*i for i in range(10)]
e_scores = [8*i for i in range(10)]

score_sums = [m_score + e_score for m_score,e_score in zip(m_scores,e_scores)]
print(score_sums)
```
