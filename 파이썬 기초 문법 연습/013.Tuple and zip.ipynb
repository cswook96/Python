## Tuple(immutable object)

```python
scores = (10,20,30,40)
print(scores)
print(scores[0])
print(scores[1])

scores[0] = 100  # Tuple은 원소를 수정할수 없다.
```

## Tuple Packing & Unpacking 

```python
# --- Packing
scores = 10,20
print('scores: ',scores)
# ---

print('-------------')

# --- Unpacking
score1,score2 = (10,20)
print('score1: ',score1)
print('score2: ',score2)
# ---
```

## for + zip

```python
m_scores = [10,20,30]
e_scores = [30,40,50]

# --- index를 통해 접근
for data_idx in range(len(m_scores)):
    m_score = m_scores[data_idx]
    e_score = e_scores[data_idx]
    print('m_score:',m_score,'e_score:',e_score)
# ---

print('-------------------')

# --- zip을 통해 접근(Packing)
for scores in zip(m_scores,e_scores):
    print('scores: ',scores)
    print('scores[0]:',scores[0])
    print('scores[1]:',scores[1])
# ---

print('-------------------')

# --- ip을 통해 접근(Unpacking)
for score1,score2 in zip(m_scores,e_scores):
    print('scrore1:',score1,'score2:',score2)
# ---
```

## Manhattan Distance

```python
v1 = [1,3,5,2,1,5,2]
v2 = [2,3,1,5,2,1,3]

m_distance = 0
for v1_val,v2_val in zip(v1,v2):
    sub = v1_val - v2_val
    
    if sub < 0:
        m_distance += -sub
    else:
        m_distance += sub

print('Manhattan Distance: ',m_distance)
```
