## Hadamard Product

```python
vectors = [[1,4,16],
           [2,8,32],
           [3,12,36],
           [4,16,64]]

h_prod = []

for dim_data in vectors:
    dim_prod = 1
    for dim_value in dim_data:
        dim_prod *= dim_value
    h_prod.append(dim_prod)

print('Hadamard Product:',h_prod)
```

## L2 Norm

```python
vectors = [[1,4,16],
           [2,8,32],
           [3,12,36],
           [4,16,64]]

n_vector = len(vectors[0])
v_norms = []

for _ in range(n_vector):
    v_norms.append(0)

for dim_data in vectors:
    for dim_idx in range(n_vector):
        v_norms[dim_idx] += dim_data[dim_idx]**2

for vec_idx in range(n_vector):
    v_norms[vec_idx] **= 0.5

print('L2 Norms: ',v_norms)
```

## Making Unit Vectors

```python
vectors = [[1,4,16],
           [2,8,32],
           [3,12,36],
           [4,16,64]]

n_dim = len(vectors)
n_vector = len(vectors[0])


# --- Unit Vetor로 만들기전 vector 크기
v_norms = []
for _ in range(n_vector):
    v_norms.append(0)

for dim_data in vectors:
    for dim_idx in range(n_vector):
        v_norms[dim_idx] += dim_data[dim_idx]**2

for vec_idx in range(n_vector):
    v_norms[vec_idx] **= 0.5

print('Unit Vector로 만들기전 벡터 크기: ',v_norms)
# ---

# --- Unit Vector 만들기
for dim_idx in range(n_dim):
    for vec_idx in range(n_vector):
        vectors[dim_idx][vec_idx] /= v_norms[vec_idx]
# ---

# --- Unit Vector로 만든후 vector 크기
v_norms = []

for _ in range(n_vector):
    v_norms.append(0)

for dim_data in vectors:
    for dim_idx in range(n_vector):
        v_norms[dim_idx] += dim_data[dim_idx]**2

for vec_idx in range(n_vector):
    v_norms[vec_idx] **= 0.5

print('Unit Vector로 만든후 벡터 크기: ',v_norms)
# ---
```

## Dot Product

```python
vectors =[[1,11],
          [2,12],
          [3,13],
          [4,14]]

dot_prod = 0

for dim_data in vectors:
    dim_prod = 1
    for dim_value in dim_data:
        dim_prod *= dim_value
    dot_prod += dim_prod

print('Dot Product: ',dot_prod)
```
