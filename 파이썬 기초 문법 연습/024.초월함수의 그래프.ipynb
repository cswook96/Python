## Sin,Cos Graphs

```python
import math
import matplotlib.pyplot as plt

PI = math.pi

low,high = 0,4*PI
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

sin = [math.sin(data) for data in linspace]
cos = [math.cos(data) for data in linspace]

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(linspace,sin,label='sin')
ax.plot(linspace,cos,label='cos')
ax.legend(fontsize=15)
ax.grid()
```

## Exp,Log Graphs

```python
import math
import matplotlib.pyplot as plt

low,high = 0.01,10
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

exp = [math.exp(data) for data in linspace]
log = [math.log(data) for data in linspace]

fig,axes = plt.subplots(2,1,figsize=(8,8))
axes[0].plot(linspace,exp)
axes[1].plot(linspace,log)
axes[0].grid()
axes[1].grid()
```

## Sigmoid Graph

```python
import math
import matplotlib.pyplot as plt

low,high = -10,10
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

sigmoid = [1/(1 + math.exp(-data)) for data in linspace]

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(linspace,sigmoid)
ax.grid()
```

## Tanh Graph

```python
import math
import matplotlib.pyplot as plt

low,high = -10,10
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

tanh = []
for data in linspace:
    sin = (math.exp(data) - math.exp(-data)) / 2
    cos = (math.exp(data) + math.exp(-data)) / 2
    tanh.append(sin/cos)

fig,ax = plt.subplots(figsize=(10,5))
ax.plot(linspace,tanh)
ax.grid()
```

## ReLU Graph

```python
import math
import matplotlib.pyplot as plt

low,high = -10,10
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

relu = [data if data > 0 else 0 for data in linspace] # [ 조건 만족 시 출력값 if 조건 else 조건 불만족 시 출력 값 for i in data] 

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(linspace,relu)
ax.grid()
```

## Information Graph

```python
import math
import matplotlib.pyplot as plt

low,high = 0.001,0.999
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

information = [math.log(1/p) for p in linspace]

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(linspace,information)
ax.set_xlabel('Probability',fontsize=20)
ax.set_ylabel('Information',fontsize=20)
ax.tick_params(labelsize=15)
ax.grid()
```

## Binary Entropy Graph

```python
import math
import matplotlib.pyplot as plt

low,high = 0.001,0.999
n_data = 100
interval = (high-low) / (n_data - 1)
linspace = [low + interval*data_idx for data_idx in range(n_data)]

binary_entropy = []
for p in linspace:
    entropy = -(p*math.log(p) + (1 - p)*math.log(1 - p))
    binary_entropy.append(entropy)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(linspace,binary_entropy)
ax.set_xlabel('Probability',fontsize=20)
ax.set_ylabel('Binary Entropy',fontsize=20)
ax.tick_params(labelsize=15)
ax.grid()
```
