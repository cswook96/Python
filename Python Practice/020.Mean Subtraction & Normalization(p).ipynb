```python
class_names = ['Korean','English','Math']
interested_data = ['sum', 'mean', 'M', 'm','M_idx','m_idx']

scores = [[10,15,30],
          [20,45,20],
          [30,35,40],
          [40,30,50]]
n_studet = len(scores)
n_class = len(scores[0])

meta_data_dict = {}
for class_name in class_names:
    meta_data_dict[class_name] = {}
    for i_data in interested_data:
        if i_data == 'M':
            meta_data_dict[class_name][i_data] = None
        elif i_data == 'm':
            meta_data_dict[class_name][i_data] = None
        else:
            meta_data_dict[class_name][i_data] = 0


for s_idx,s_scores in enumerate(scores):
    for c_idx,c_score in enumerate(s_scores):
        meta = meta_data_dict[class_names[c_idx]]
        meta['sum'] += c_score

        if meta['M'] == None or c_score > meta['M']:
            meta['M'] = c_score
            meta['M_idx'] = s_idx
        
        if meta['m'] == None or c_score < meta['m']:
            meta['m'] = c_score
            meta['m_idx'] = s_idx

for c_idx in range(n_class):
    meta = meta_data_dict[class_names[c_idx]]
    meta['mean'] = meta['sum']/n_studet

print('Meta data dictionary:')
for k, v in meta_data_dict.items():
    print(f'subject: {k} ==> {v}')

scores_ms = []
scores_nomarlization = []
for s_idx,s_scores in enumerate(scores):
    scores_ms_vec = []
    scores_nomarlization_vec = []
    
    for c_idx,c_score in enumerate(s_scores):
        meta = meta_data_dict[class_names[c_idx]]
        scores_ms_vec.append(c_score - meta['mean'])
        scores_nomarlization_vec.append((c_score-meta['m'])/(meta['M']-meta['m']))
    
    scores_ms.append(scores_ms_vec)
    scores_nomarlization.append(scores_nomarlization_vec)

print('\nMean Subtraction:\t\t Nomarlization:')
for score_ms_vec, scores_nomarlization_vec in zip(scores_ms, scores_nomarlization):
    print(score_ms_vec,'\t',scores_nomarlization_vec)
```
