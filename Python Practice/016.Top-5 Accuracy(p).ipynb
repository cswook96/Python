### Membership Test

### Sorting with Indices(내림차순)

```python
predictions = [40,25,10,80,60,20]
n_pred = len(predictions)
print('Predictions:',predictions)
print('\n')

sorted_predictions = []
sorted_predictions_indices = []

for _ in range(n_pred):
    M,M_idx = None,0
    for pred_idx,pred in enumerate(predictions):
        if pred_idx in sorted_predictions_indices:     # 그전에 이미 최댓값으로 구한값은 무시하고 나머지 값들로 최댓값 구하기
            pass
        elif M == None or pred > M:
            M = pred
            M_idx = pred_idx
    
    sorted_predictions.append(M)
    sorted_predictions_indices.append(M_idx)

print('Sorted predictions:',sorted_predictions)
print('Sorted prediction indices:',sorted_predictions_indices)
```

## Top-5 Accuracy
Classification에서 정확한 예측이 아니더라도, 유의미한 예측일때 측정하는 측정값

```python
prediction = [40,25,10,80,60,20,5,30,10,15] # 각 클래스에대한 예측 점수
label = 0

# --- 예측 점수를 확률로 바꾸기(모든 확률의 합은 1)
sum = 0
for pred in prediction:
    sum += pred

pred_probs = []
for pred in prediction:
    pred_probs.append(pred/sum)
print('Prediction Probabilities :\n',pred_probs,'\n')
# ---

# --- Sorting with Indices(내림차순)
sorted_probs = []
sorted_prob_indices = []

for _ in range(len(pred_probs)):
    M,M_idx = None,0
    for prob_idx,prob in enumerate(pred_probs):
        if prob_idx in sorted_prob_indices:
            pass
        elif M == None or prob > M:
            M = prob
            M_idx = prob_idx
    sorted_probs.append(M)
    sorted_prob_indices.append(M_idx)

print('Sorted probilities:\n',sorted_probs)
print('Sorted probability indices:\n',sorted_prob_indices,'\n')
# ---

top_5_pred = sorted_prob_indices[:5]    # Top-5 prediction 구하기
print(f'Label: {label}, Top-5 prediction: {top_5_pred}')
if label in top_5_pred:
    print('Correct prediction!')
else:
    print('Wrong prediction!')
```
