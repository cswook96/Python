## Dictionary 만들기 & Key로 value 접근하기

```python
char_cnt_dict = {'a':3,'b':2,'c':10,'d':7}

print(char_cnt_dict,'\n')

print(char_cnt_dict['a'])
print(char_cnt_dict['b'])
print(char_cnt_dict['c'])
print(char_cnt_dict['d'])
```

## Key, Value 추가하기

```python
test_dict = dict()
print(test_dict)

test_dict['a']  = 1
print(test_dict)

test_dict['b']  = 2
print(test_dict)

test_dict['c']  = 3
print(test_dict)

test_dict['a']  = 10
print(test_dict)
```

## keyError와 dict.get
- KeyError: 참조하려는 key가 없을 때 발생
- dict.get: 참조하려는 key값이 없을때 설정한 초깃값을 return

```python
test_dict = {}

print(test_dict.get('a',0))
print(test_dict.get('a','사전에 참조하려는 key값이 없습니다'))
```

## Dictionary로 글자수 세기

```python
test_string = 'asffdhkj34rmsnasdwe2!3##$MsdasgAA'

cnt_dict = {}
for string in test_string:
    cnt_dict[string] = cnt_dict.get(string,0) + 1

print(f'글자수:\n{cnt_dict}')
```

## One-hot Encoding

```python
labels = ['dog','cat','dolphin','tiger','dog','cat','cow']
onehot_enc_dict = {'dog':0,'cat':1,'tiger':2,'dolphin':3,'rabbit':4,'cow':5}
n_label = len(labels)
n_class = len(onehot_enc_dict)

onehot_mat = []
for label in labels:
    onehot_vec = [0]*n_class
    onehot_vec[onehot_enc_dict[label]] = 1
    onehot_mat.append(onehot_vec)

for onehot_vec in onehot_mat:
    print(onehot_vec)
```
