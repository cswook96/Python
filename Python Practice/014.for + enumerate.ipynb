## 수학 장학생 선정

```python
# 조건: 수학 90점 이상이면서 동시에 영어,국어 70점 이상

m_scores = [92,60,80,94,50,60,95]
e_scores = [70,40,50,80,70,40,80]
k_scores = [30,20,60,80,30,60,65]

for student_idx, m_score in enumerate(m_scores):
    if m_score >= 90:
        e_score = e_scores[student_idx]
        k_score = k_scores[student_idx]
        if e_score >= 70 and k_score >= 70:
            print('장학생 index:',student_idx,'장학생 (수학,영어,국어) 점수:',(m_score,e_score,k_score))
```

## 최댓값, 최솟값, 그 값들의 index



```python
scores = [50,30,90,70,30,95,40]

M,m = scores[0],scores[0]
M_idx, m_idx = 0,0

for score_idx,score in enumerate(scores):
    if score > M:
        M = score
        M_idx = score_idx
    
    if score < m:
        m = score
        m_idx = score_idx

print('Max index:',M_idx,'Max Value:',M,' min index:',m_idx,'min Value:',m)
```

## One-hot-Encoding

```python
labels = ['dog','cat','dolphin','tiger'] # dolphin:0 ,cat:1,dog:2,tiger:3
print('Before one-hot-Encoding:')
print(labels)

n_label = len(labels)
n_class = 4

one_hot_mat = []
for _ in range(n_label):
    one_hot_vec = []
    for _ in range(n_class):
        one_hot_vec.append(0)
    one_hot_mat.append(one_hot_vec)

for label_idx,label in enumerate(labels):
    one_hot_vec = one_hot_mat[label_idx]
    if label == 'dolphin':
        one_hot_vec[0] = 1
    elif label == 'cat':
        one_hot_vec[1] = 1
    elif label == 'dog':
        one_hot_vec[2] = 1
    elif label == 'tiger':
        one_hot_vec[3] = 1
    else:
        print('존재하지 않는 클래스 입니다.')

print('\n')

print('After one-hot-Encoding:')
print(one_hot_mat)
```
