```python
df.columns = [x.lower().strip().replace(' ','_') for x in df.columns]
```

```python
def miss_data(df):
    total = df.isnull().sum()
    percent = (df.isnull().sum()/df.isnull().count()*100)
    miss_df = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    
    dtypes = []
    for col in df.columns:
        dtypes.append(str(df[col].dtype))
    miss_df['Types'] = dtypes
    return np.transpose(miss_df)
```

```python
miss_data(df)
```
