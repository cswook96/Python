```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/innerwave/연습/data/student-mat.csv')
df
```

```python
df.info()
```

```python
df['school'] = df['school'].replace({'GP':0, 'MS':1}).astype('category')
```

```python
df['sex'] = df['sex'].replace({'F':0, 'M':1}).astype('category')
```

```python
df['address'] = df['address'].replace({'U':0, 'R':1}).astype('category')
```

```python
df['famsize'] = df['famsize'].replace({'GT3':0, 'LE3':1}).astype('category')
```

```python
df['Pstatus'] = df['Pstatus'].replace({'A':0, 'T':1}).astype('category')
```

```python
df[['Medu', 'Fedu']] = df[['Medu', 'Fedu']].astype('category')
```

```python
df['Mjob'] = df['Mjob'].replace({'at_home':0, 'health':1, 'other':2, 'services':3, ' teacher':4}).astype('category')
```

```python
df['Mjob'] = df['Mjob'].replace({'teacher':4}).astype('category')
```

```python
df['Fjob'] = df['Fjob'].replace({'at_home':0, 'health':1, 'other':2, 'services':3, 'teacher':4}).astype('category')
```

```python
df['reason'] = df['reason'].replace({'course':0, 'other':1, 'home':2, 'reputation':3}).astype('category')
```

```python
df['guardian'] = df['guardian'].replace({'mother':0, 'father':1, 'other':2}).astype('category')
```

```python
df.loc[:, 'schoolsup': 'romantic'] = df.loc[:, 'schoolsup': 'romantic'].replace({'no':0, 'yes':1}).astype('category')
```

```python
df.loc[:, 'G1':'G3'] = df.loc[:, 'G1':'G3'].astype('category')
```

```python
df
```

```python
df.info()
```

```python
x_train = df.iloc[:int(len(df)*0.8), :-1]
y_train = df.iloc[:int(len(df)*0.8), -1]
x_valid = df.iloc[int(len(df)*0.8):, :-1]
y_valid = df.iloc[int(len(df)*0.8):, -1]
```

```python
rf_model = RandomForestClassifier(n_estimators=70, n_jobs=-1, max_depth=3)
```

```python
rf_model.get_params()
```

```python
rf_model.fit(x_train, y_train)
```

```python
prediction = rf_model.predict(x_valid)
```

```python
accuracy_score(y_valid, prediction)
```
